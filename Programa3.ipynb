```

def pagotrabajador(horas,costoh=150):
    dia = horas * costoh
    print("El pago de un dia de trabajo es: ",dia)
    semana = 5 * dia
    print("El pago en una semana de trabajo es: ",semana)
    mes = 20 * dia
    print("El pago en un mes de trabajo es: ",mes)

a= float(input("Numero de horas : "))
b= float(input("Pago por hora : "))

pagotrabajador(a)

pagotrabajador(a,b)
```
